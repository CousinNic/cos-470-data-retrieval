```python
from post_parser_record import PostParserRecord
post_reader = PostParserRecord("Posts.xml")
import numpy as np 
from matplotlib import pyplot as plt
```

```python
import re, string
index = {}
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#updates index based of text from current document
def update_index(text,post_id):
  tokens = word_tokenize(re.sub(r'[^a-z]', ' ', text))
  for word in tokens:
    if word in index:
      if post_id not in index[word]:
        index[word].add(post_id)
    else:
        index[word] = set([post_id])

#indexes questions
print("Indexing the questions...")
for question_id in post_reader.map_questions:
  text = (post_reader.map_questions[question_id].title + " " + post_reader.map_questions[question_id].body)
  update_index(text.lower(),question_id)  
print("done")

#indexes answers
print("Indexing the answers...")
for answer_id in post_reader.map_just_answers:
  text = post_reader.map_just_answers[answer_id].body
  update_index(text.lower(),answer_id)
print("done")
```
